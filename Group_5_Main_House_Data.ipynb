**ANOVA**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
mod = ols('price~sqft_lot', data=df).fit()
aov = sm.stats.anova_lm(mod, type=2)
print(aov)
```

```python
#two way anova
mod1 = ols('price~sqft_lot+sqft_living', data=df).fit()
aov1 = sm.stats.anova_lm(mod1, type=2)
print(aov1)
```

**Ridge Regression**


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

y_data = df['bathrooms']
x_data=df.drop('bathrooms',axis=1)


x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.45, random_state=0)
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train[['bedrooms', 'sqft_living', 'sqft_lot','floors','waterfront']])
x_test_pr=pr.fit_transform(x_test[['bedrooms', 'sqft_living', 'sqft_lot','floors','waterfront']])
```

```python
df.head()
```

```python
from sklearn.linear_model import Ridge
```

```python
RigeModel=Ridge(alpha=1)
```

```python
RigeModel.fit(x_train_pr, y_train)
```

```python
yhat = RigeModel.predict(x_test_pr)
```

```python
print('predicted:', yhat[0:4])
print('test set :', y_test[0:4].values)
```

```python
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```

```python
RigeModel = Ridge(alpha=10) 
RigeModel.fit(x_train_pr, y_train)
RigeModel.score(x_test_pr, y_test)
```
