# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['sqft_living']]
Y = df['price']

lm = LinearRegression()
lm.fit(X,Y)
```

```python
Yhat=lm.predict(X)
Yhat[0:5]
```

```python
lm.intercept_
```

```python
lm.coef_
```

Equation: price = -43580.74 + 280.62 x **sqft_living**

# Multiple Linear regression

```python
X = df[['sqft_living', 'sqft_above', 'grade']]
Y = df['price']

lm1 = LinearRegression()
lm1.fit(X,Y)
```

```python
lm1.intercept_
```

```python
lm1.coef_
```

Equation: price = -651195.12 + 2.35 x **sqft_living** - 7.77 x **sqft_above** + 1.01 x **grade**
