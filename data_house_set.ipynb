```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv("Cleaned_kc_house_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Value counts is a good way of understanding how many units of each characteristic/variable we have. We can apply the "value_counts" method on the column "bedrooms".
df['bedrooms'].value_counts()
```

```python
bedrooms_counts = df['bedrooms'].value_counts().to_frame()
bedrooms_counts.rename(columns={'bedrooms': 'value_counts'}, inplace=True)
bedrooms_counts.index.name = 'bedrooms'
bedrooms_counts
```

```python
#We can repeat the above process for the variable 'floors'.
floors_counts = df['floors'].value_counts().to_frame()
floors_counts.rename(columns={'floors': 'value_counts'}, inplace=True)
floors_counts.index.name = 'floors'
floors_counts
```

```python
#The "groupby" method groups data by different categories. The data is grouped based on one or several variables, and analysis is performed on the individual groups.
#For example, let's group by the variable "floors". We see that there are 6 different categories of floors
df['floors'].unique()
```

```python
#We can select the columns 'floors', 'sqft_lot' and 'price', then assign it to the variable "df_group_one".
df_group_one = df[['floors', 'sqft_lot', 'price']]

#We can then calculate the average price for each of the different categories of data.
# grouping results
df_group_one = df_group_one.groupby(['floors'],as_index=False).mean()
df_group_one
```

```python
#You can also group by multiple variables. For example, let's group by both 'floors' and 'bedrooms'
# grouping results
df_gptest = df[['floors', 'bedrooms', 'price']]
grouped_test1 = df_gptest.groupby(['floors', 'bedrooms'],as_index=False).mean()
grouped_test1
```

```python
#This grouped data is much easier to visualize when it is made into a pivot table. A pivot table is like an Excel spreadsheet, with one variable along the column and another along the row. We can convert the dataframe to a pivot table using the method "pivot" to create a pivot table from the groups
grouped_pivot = grouped_test1.pivot(index='floors',columns='bedrooms')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
#Sometimes we would like to know the significant of the correlation estimate.
#p-value is  <  0.001: we say there is strong evidence that the correlation is significant.
#the p-value is  <  0.05: there is moderate evidence that the correlation is significant.
#the p-value is  <  0.1: there is weak evidence that the correlation is significant.
#the p-value is  >  0.1: there is no evidence that the correlation is significant.

from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df['bedrooms'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

```python
pearson_coef, p_value = stats.pearsonr(df['floors'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```
