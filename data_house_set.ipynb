```python
import pandas as pd
import numpy as np 
```

```python
#Value counts is a good way of understanding how many units of each characteristic/variable we have. We can apply the "value_counts" method on the column "bedrooms".
df['bedrooms'].value_counts()
```

```python
bedrooms_counts = df['bedrooms'].value_counts().to_frame()
bedrooms_counts.rename(columns={'bedrooms': 'value_counts'}, inplace=True)
bedrooms_counts.index.name = 'bedrooms'
bedrooms_counts
```

```python
#We can repeat the above process for the variable 'floors'.
floors_counts = df['floors'].value_counts().to_frame()
floors_counts.rename(columns={'floors': 'value_counts'}, inplace=True)
floors_counts.index.name = 'floors'
floors_counts
```

```python
#The "groupby" method groups data by different categories. The data is grouped based on one or several variables, and analysis is performed on the individual groups.
#For example, let's group by the variable "floors". We see that there are 6 different categories of floors
df['floors'].unique()
```

```python
#We can select the columns 'floors', 'sqft_lot' and 'price', then assign it to the variable "df_group_one".
df_group_one = df[['floors', 'sqft_lot', 'price']]

#We can then calculate the average price for each of the different categories of data.
# grouping results
df_group_one = df_group_one.groupby(['floors'],as_index=False).mean()
df_group_one
```

```python
#You can also group by multiple variables. For example, let's group by both 'floors' and 'bedrooms'
# grouping results
df_gptest = df[['floors', 'bedrooms', 'price']]
grouped_test1 = df_gptest.groupby(['floors', 'bedrooms'],as_index=False).mean()
grouped_test1
```
