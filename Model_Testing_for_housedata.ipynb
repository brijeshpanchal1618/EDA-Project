```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = np.array(df['grade']).reshape(-1,1)
y = np.array(df['price'])


pre_process = PolynomialFeatures(degree=2, include_bias=False)

x_poly = pre_process.fit_transform(x)

pr_model = LinearRegression().fit(x_poly, y)

y_pred = pr_model.predict(x_poly)
```

```python
r_sq = pr_model.score(x_poly, y)
print(r_sq)
print(pr_model.intercept_)
print(pr_model.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
x_train_a = np.array(x_train).reshape(-1,1)
y_train_a = np.array(y_train)

x_test_a = np.array(x_test).reshape(-1,1)
y_test_a = np.array(y_test)


lr = LinearRegression()
lr.fit(x_train_a, y_train_a)
y_pred = lr.predict(x_test_a)
print(r2_score(y_test_a, y_pred))
```

```python
#Linear regression for trained data 
plt.plot(x_train, lr.predict(x_train_a), color="r")
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# applying polynomial regression degree 2
poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_trans = poly.fit_transform(x_train_a)
x_test_trans = poly.transform(x_test_a)
#include bias parameter
lr = LinearRegression()
lr.fit(x_train_trans, y_train_a)
y_pred = lr.predict(x_test_trans)
print(r2_score(y_test_a, y_pred))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
# Regression Plot for Grade and Price
X_new = np.linspace(1,13).reshape(50, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(x_test, y_test, "g.",label='Testing points')
plt.xlabel("grade")
plt.ylabel("price")
plt.legend()
plt.show()
```
